**FEATURE SELECTION**

```python
X=df.drop(['stress_level'],axis=1)
X
```

```python
Y=df['stress_level']
Y
```

**SPLITTING DATA**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

**STANDARDIZING DATA**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
X_train=pd.DataFrame(X_train,columns=X.columns)
X_test=pd.DataFrame(X_test,columns=X.columns)
```

**MODELING DATA**

```python
#IMPORTING LIBRARIES
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
#RANDOM FOREST CLASSIFIER
rf=RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=3)
rf.fit(X_train,Y_train)
rf_pred=rf.predict(X_test)
```

```python
#K-NEAREST NEIGHBOR CLKASSIFIER
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
knn_pred = knn.predict(X_test)
```

```python
#DECISION TREE CLASSIFIER
DT=DecisionTreeClassifier(criterion='gini',splitter='best',max_features=3,max_depth=3)
DT.fit(X_train,Y_train)
DecisionTreeClassifier(max_depth=3,max_features=3)
DT_pred=DT.predict(X_test)
```

```python
#SUPPORT VECTOR MACHINE CLASSIFIER
svm = SVC()
svm.fit(X_train, Y_train)
svm_pred = svm.predict(X_test)
```

```python
#LOGISTIC REGRESSION CLASSIFIER
log=LogisticRegression()
log.fit(X_train,Y_train)
log_pred=log.predict(X_test)
```

```python
#NAIVE BAYES CLASSIFICATION
naive_bayes = GaussianNB()
naive_bayes.fit(X_train, Y_train)
nb_pred = naive_bayes.predict(X_test)
```

**MODEL EVALUATION**

```python
from sklearn.metrics import accuracy_score
print("ACCURACY SCORES\n")
print("RANDOM FOREST :",accuracy_score(Y_test,rf_pred))
print("KNN :",accuracy_score(Y_test, knn_pred))
print("DECISION TREE :",accuracy_score(Y_test,DT_pred))
print("SVM Accuracy:", accuracy_score(Y_test, svm_pred))
print("logistic regression :", accuracy_score(Y_test, log_pred))
print("naive bayes :", accuracy_score(Y_test, nb_pred))
```

```python
from sklearn.metrics import confusion_matrix
print("CONFUSION MATRICES\n")
print("RANDOM FOREST :\n",confusion_matrix(Y_test,rf_pred))
print("KNN :\n",confusion_matrix(Y_test, knn_pred))
print("DECISION TREE :\n",confusion_matrix(Y_test,DT_pred))
print("SVM :\n", confusion_matrix(Y_test, svm_pred))
print("logistic regression :\n",confusion_matrix(Y_test, log_pred))
print("naive bayes:\n", confusion_matrix(Y_test, nb_pred))
```

```python
from sklearn.metrics import precision_score
print("PRECISION-SCORES\n")
print("RANDOM FOREST :",precision_score(Y_test,rf_pred,average='weighted'))
print("KNN :",precision_score(Y_test, knn_pred, average='weighted'))
print("DECISION TREE :",precision_score(Y_test,DT_pred, average='weighted'))
print("SVM :",precision_score(Y_test, svm_pred, average='weighted'))
print("logistic regression :",precision_score(Y_test, log_pred, average='weighted'))
print("naive bayes:",precision_score(Y_test, nb_pred, average='weighted'))
```

```python
from sklearn.metrics import recall_score
print("RECALL-SCORES\n")
print("RANDOM FOREST :",recall_score(Y_test,rf_pred, average='weighted'))
print("KNN :",recall_score(Y_test, knn_pred, average='weighted'))
print("DECISION TREE :",recall_score(Y_test,DT_pred, average='weighted'))
print("SVM :", recall_score(Y_test, svm_pred, average='weighted'))
print("logistic regression :",recall_score(Y_test, log_pred, average='weighted'))
print("naive bayes :", recall_score(Y_test, nb_pred, average='weighted'))
```

```python
from sklearn.metrics import f1_score
print("F1-SCORES\n")
print("RANDOM FOREST :",f1_score(Y_test,rf_pred, average='weighted'))
print("KNN :",f1_score(Y_test, knn_pred, average='weighted'))
print("DECISION TREE :",f1_score(Y_test,DT_pred, average='weighted'))
print("SVM :",f1_score(Y_test, svm_pred, average='weighted'))
print("logistic regression :",f1_score(Y_test, log_pred, average='weighted'))
print("naive bayes :",f1_score(Y_test, nb_pred, average='weighted'))
```
